```python
import pandas as pd 
import numpy as np 
import ast
```

```python
data = pd.read_csv("movies_metadata.csv")
data.head(1)
```

```python
# genres
# imdb_id 
# overview
# poster_path
# title
# id

movies = data[[ "id","title", "genres", "imdb_id", "overview", "poster_path", ]]
movies.head()
```

```python
movies.id  = pd.to_numeric(movies.id , errors='coerce')
```

```python
movies.id.isnull().sum()
```

```python
movies = movies.dropna()
```

```python
credits = pd.read_csv("credits.csv")
```

```python
movies = movies.merge(credits, on= "id")
```

```python
movies.head(3)
```

# Now
 
 Three columns here have very complicated format and its very to hard to read to us human and kinda impossible for computers,
 
 lets change all three of them to a list of some usefull tags and then we'll combine them. 

```python
def change_genres(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
'''  applying to genres column  '''
movies.genres= movies.genres.apply(change_genres)
```

```python
movies.head(1)
```

```python
#### '''   creating function for cast cvolumns    '''
movies.cast = movies.cast.apply(change_genres)
```

```python
movies.head(2)
```

```python
def change_crew(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
    return l

movies.crew= movies.crew.apply(change_crew)
```

```python
movies.head(2)
```

```python
movies.overview = movies.overview.apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies.genres = movies.genres.apply(lambda x: [ i.replace(" ", "") for i in x])
movies.cast = movies.cast.apply(lambda x: [ i.replace(" ", "") for i in x])
movies.crew = movies.crew.apply(lambda x: [ i.replace(" ", "") for i in x])
```

```python
movies["tags"] = movies.overview + movies.cast + movies.crew + movies.genres
```

```python
movies.head()
```

```python
new_data = movies[['id','title','tags']]
```

```python
new_data
```

```python
new_data.tags = new_data.tags.apply(lambda x: " ".join(x))
```

```python
new_data.tags = new_data.tags.apply(lambda x: x.lower())
```

```python
new_df = new_data.copy()
```

```python
new_data = new_df.iloc[:8500,:]
```

```python
new_data
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
stemed = new_data.tags.apply(stem)
```

```python
new_data.tags = stemed
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_data.tags).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar_score = cosine_similarity(vectors)
```

```python
similar_score
```

```python
def recommend(movie):
    movie_index = new_data[new_data.title == movie].index[0]
    distance_of_the_selected_movie = similar_score[movie_index]
    movie_list = sorted(list(enumerate(distance_of_the_selected_movie)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_data.iloc[i[0]].id)
```

```python
recommend("Toy Story")
```

```python
import pickle 
pickle.dump(new_data.to_dict(), open("movie_dict.pkl", "wb"))
```
